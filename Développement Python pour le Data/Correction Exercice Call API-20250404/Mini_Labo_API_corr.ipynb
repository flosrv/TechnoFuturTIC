Nous allons donc pouvoir aller chercher les données météos sur base des coordonnées de nos communes

```python
# Récupération des coordonnées pour utiliser l'API open météo

coordonnees = dict(zip(df['lat'],df['lon']))
```

```python
### example:
### doc url: https://open-meteo.com/
### here : max 10_000 call per IP address, because it's free: don't be abusive 


### built url from documentation: 
### try in browser -->> return json
url = "https://archive-api.open-meteo.com/v1/archive?latitude=52.52&longitude=13.41&start_date=2025-02-25&end_date=2025-03-11&hourly=temperature_2m,wind_speed_10m,rain,snowfall"


### the values associated to "latitude" and "longitude" are parameters => changeable with for loop ! 
```

```python
# Script call API avec une boucle sur les coordonnées.

with req.Session() as s:

    for lat,lon in coordonnees.items():
            fill_url = f'https://historical-forecast-api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&start_date=2025-02-27&end_date=2025-03-13&hourly=temperature_2m,rain,wind_speed_10m'
            r = s.get(fill_url)
            r = r.json()
            df_dict = {'time':r.get('hourly').get('time'),
                    'temperature_2m':r.get('hourly').get('temperature_2m'),
                    'wind_speed_10m':r.get('hourly').get('wind_speed_10m'),
                    'rain':r.get('hourly').get('rain')
                    }
            ready_to_merge = pd.DataFrame(df_dict)
            ready_to_merge[['latitude','longitude']] = [lat,lon]
            ready_to_merge['time'] = pd.to_datetime(ready_to_merge['time'])
            ready_to_merge = ready_to_merge.merge(df,how='left',left_on=['latitude','longitude'],right_on=['lat','lon'])
            full_data = ready_to_merge[['time','temperature_2m','wind_speed_10m','rain','index']]
            full_data = full_data.rename(columns={'index':'commune_id'})
            full_data.to_sql(name='meteo_data',con=con_string_insert,if_exists='append',index=False)    
```
