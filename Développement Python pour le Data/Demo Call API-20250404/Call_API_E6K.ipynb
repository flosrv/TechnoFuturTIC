```python
import pandas as pd 
import numpy as np 

import requests as rq
from sqlalchemy import create_engine,text,Float,Integer,String
```

```python
url = 'https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&hourly=temperature_2m'
```

```python
response = rq.get(url)
response.status_code
```

```python
response.headers
```

```python
rawdata = response.json()
rawdata
```

```python
rawdata.keys()
```

```python
rawdata.get('hourly')
```

```python
rawdata.get('hourly').get('time')
```

```python
datadict = {'Time':rawdata.get('hourly').get('time'),
            'Temperature_2m':rawdata.get('hourly').get('temperature_2m')}

df = pd.DataFrame(datadict)
df
```

```python
latitude = rawdata.get('latitude')
longitude = rawdata.get('longitude')

df['lat'] = latitude
df['lon'] = longitude
```

```python
df
```

```python
dbms = 'mssql+pyodbc'
server_name = r'GEORGES\DATA'
driver = 'ODBC+driver+17+for+SQL+Server'

create_db_string = f'{dbms}://{server_name}/master?trusted_connection=yes&driver={driver}'
```

```python
create_db_engine = create_engine(create_db_string,isolation_level='AUTOCOMMIT')
```

```python
create_query = """
IF NOT EXISTS (SELECT [name] FROM sys.databases
				WHERE [name] = 'meteo')
                     BEGIN
                        CREATE DATABASE meteo
                     END

"""
```

```python
create_db_engine.connect()
create_db_engine.connect().execute()
create_db_engine.dispose()
```

```python
with create_db_engine.connect() as conn :
    conn.execute(text(create_query))

```

```python
dbms = 'mssql+pyodbc'
server_name = r'GEORGES\DATA'
driver = 'ODBC+driver+17+for+SQL+Server'
database = 'meteoe6k'

insert_string = f'{dbms}://{server_name}/{database}?trusted_connection=yes&driver={driver}'
```

```python
insert_engine = create_engine(insert_string)
```

```python
typedict = {'Time': String(50),
            'Temperature_2m':Float(),
            'lat':Float(),
            'lon':Float()}
```

```python
df.to_sql(name="temperature",con=insert_engine,if_exists='replace',index=False,dtype=typedict)
```

```python
retrieval_query = "SELECT * FROM temperature"
```

```python
retrieved_data = pd.read_sql(retrieval_query,con=insert_engine)
retrieved_data
```
