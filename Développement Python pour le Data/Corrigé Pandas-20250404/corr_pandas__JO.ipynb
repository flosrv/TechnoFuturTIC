# .info cusom:

```python
def INFO(data):
    res = pd.concat([
        data.dtypes
        ,data.isnull().sum()
        ,data.nunique()
    ],axis=1)
    res.columns = ['dtype','isnull_sum','nunique']
    
    return res
```
