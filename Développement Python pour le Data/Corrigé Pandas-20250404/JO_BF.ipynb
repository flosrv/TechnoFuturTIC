31) Quels sont les pays gagnant les plus mÃ©daillÃ©s par sport distinct 

```python
GB = (df
      .loc[:,['NOC','Sport','Medal']]
      .groupby(['NOC','Sport'])
      .count()
      .sort_values(by=['Sport','Medal'],ascending=[True,False])
      .reset_index()
     )
GB.drop_duplicates(subset=['Sport']).sort_values(by='Medal',ascending=False)
```

32) Quels Ã©taient les pays les plus efficaces durant la guerre froide de 1952 Ã  1988 (i.e. le pays 
avec le plus de mÃ©dailles par nombre dâ€™athlÃ¨tes envoyÃ©s)

```python
df_medal = df.copy()
cold_war = df_medal['Year'].between(1952,1988)
df_medal = df_medal[cold_war]
df_medal.loc[~df_medal['Medal'].isna(),'got_medal'] = 1
df_medal.loc[df_medal['Medal'].isna(),'got_medal'] = 0
df_medal = df_medal.groupby(['Year','NOC'],as_index=False).agg({'Name':'count',
                                      'got_medal':'sum'})
df_medal['eff'] = df_medal['got_medal']/df_medal['Name']
df_medal.sort_values(by='eff',ascending=False)
```

```python
eff_df = df.loc[cold_war,['NOC','ID','Medal']].groupby('NOC').count()\
                .sort_values(by='Medal',ascending=False)
eff_df['eff'] = eff_df['Medal']/eff_df['ID']
eff_df.sort_values(by='eff',ascending=False)
```
